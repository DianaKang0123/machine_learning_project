```python
import pandas as pd
from datetime import datetime

# 'last_review' 열을 datetime 형식으로 변환
a_df['last_review'] = pd.to_datetime(a_df['last_review'])

# 가장 최근 날짜 찾기
latest_review_date = a_df['last_review'].min()

# 결과 출력
print("가장 최근 날짜:", latest_review_date)
```

```python
pre_a_df = a_df.drop(labels = ['last_review', 'id','name','host_id','host_name','neighbourhood'], axis=1)
```

```python
pre_a_df.reviews_per_month.value_counts()
```

```python
pre_a_df.loc[:,'reviews_per_month']= pre_a_df.reviews_per_month.fillna(0)
```

```python
pre_a_df.isna().sum()
```

```python
pre_a_df.loc[:,'target'] =  pre_a_df.price
```

```python
pre_a_df = pre_a_df.drop(labels = ['price'], axis=1)
```

```python
pre_a_df
```

```python
pre_a_df.neighbourhood_group.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
result = label_encoder.fit_transform(pre_a_df['neighbourhood_group'])
pre_a_df.neighbourhood_group = result

print(result) 
```

```python
pre_a_df.room_type.value_counts()
```

```python
def room_rating(room_type):
    if room_type == 'Entire home/apt':
        return 2
    elif room_type == 'Private room':
        return 1
    elif room_type == 'Shared room':
        return 0
    else:
        return None  # 알 수 없는 경우 0을 반환합니다.
```

```python
pre_a_df.loc[:,'room_type'] = pre_a_df.room_type.apply(room_rating)
```

```python
pre_a_df.room_type.value_counts()
```

```python
pre_a_df.describe().T
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = pre_a_df.iloc[:, :-1], pre_a_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)
```

```python
train_df = pd.concat([X_train, y_train], axis =1).reset_index(drop=True)
train_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = train_df.iloc[:, :-1], train_df.iloc[:, -1]

X_train, X_val, y_train, y_val = \
train_test_split(features, targets, test_size=0.2, random_state=124)
```

```python
l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f},, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```
