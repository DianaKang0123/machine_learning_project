## 연속형 데이터

```python
# object 타입을 가진 칼럼들을 제외한 칼럼들 선택
pre_p_num_df = pre_p_df[pre_p_df.columns[~pre_p_df.columns.isin(object_columns)]]
```

```python
pre_p_num_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = pre_p_num_df.iloc[:, :-1],pre_p_num_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))

    # 데이터프레임 생성
    evaluation_df = pd.DataFrame({
        'MSE':[MSE], 
        'RMSE':[RMSE], 
        # 'MSLE':[MSLE],
        # 'RMSLE':[RMSLE],
        'R2':[R2],
    })

    return evaluation_df
```

```python
prediction = l_r.predict(X_test)
first_lr_test = get_evaluation(y_test, prediction)
```

```python
prediction = l_r.predict(X_val)
first_lr_val = get_evaluation(y_val, prediction)
```

```python
prediction = l_r.predict(X_test)
first_poly_test = get_evaluation(y_test, prediction)
```

```python
prediction = l_r.predict(X_val)
first_poly_val = get_evaluation(y_val, prediction)
```

```python
def classify_car_size(body_type):
    if body_type in ['SUV', 'Crossover', 'Compact SUV']:
        return 'Large'
    elif body_type in ['Sedan', 'Compact sedan', 'Coupe', 'Station Wagon']:
        return 'Medium'
    elif body_type in ['Hatchback', 'Compact hatchback']:
        return 'Small'
    elif body_type in ['Mini Van', 'Van', 'MPV', 'Micro Van', 'Double Cabin', 'Pick Up', 'Truck', 'Single Cabin']:
        return 'Van/Truck'
    else:
        return 'Other'
```

```python
pre_p_df.loc[:,'body'] =pre_p_df['body'].apply(classify_car_size)
```

```python
pre_p_df.body.value_counts()
```

```python
def body_grade(body):
    if body == 'Medium':
        return 1
    elif body == 'Small':
        return 2
    elif body == 'Large':
        return 3
    elif body == 'Van/Truck':
        return 4
    else:
        return 0
```

```python
pre_p_df.loc[:,'body'] =pre_p_df['body'].apply(body_grade)
```

```python
pre_p_df.body = pre_p_df.body.astype(int)
```
