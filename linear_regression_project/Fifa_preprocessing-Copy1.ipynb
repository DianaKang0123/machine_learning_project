```python
pre_f_df = pre_f_df.drop(labels=['ID','Photo', 'Flag','Club Logo'], axis=1)
```

```python
def calc(currency_string):
    currency_string = currency_string.replace('¢æ', '')  # 통화 기호 제거
    if 'M' in currency_string:
        return float(currency_string.replace('M', '')) * 100000  # M일 경우 백만을 곱함
    elif 'K' in currency_string:
        return float(currency_string.replace('K', '')) * 1000  # K일 경우 천을 곱함
    else:
        return float(currency_string)
```

```python
pre_f_df.loc[:,'Value'] = pre_f_df.Value.apply(calc)
```

```python
pre_f_df.loc[:,'Wage'] = pre_f_df.Wage.apply(calc)
```

```python
pre_f_df.loc[:,'Release Clause'] = pre_f_df['Release Clause'].apply(calc)
```

```python
pre_f_df = pre_f_df.astype({'Value': float, 'Wage': float, 'Release Clause': float})
pre_f_df
```

```python
# 범주형 데이터 칼럼
object_columns = pre_f_df.select_dtypes(include=['object'])

# 결과 출력
display(object_columns)
```

```python
def height(height):
    # 피트와 인치를 분리
    feet, inches = map(int, height.split("'"))
    # 피트를 센티미터로 변환
    feet_cm = feet * 30.48
    # 인치를 센티미터로 변환
    inches_cm = inches * 2.54
    # 두 값의 합을 반환
    return feet_cm + inches_cm
```

```python
pre_f_df['Height'] = pre_f_df['Height'].apply(height).astype(float)
```

```python
def remove_unit(weight):
    # 숫자만 담을 변수 초기화
    numbers = ''
    # 문자열을 순회하면서 숫자인 부분만 추출
    for char in weight:
        if char.isdigit():
            numbers += char
    return numbers
```

```python
pre_f_df['Weight'] = pre_f_df['Weight'].apply(remove_unit).astype(float)
```

```python
# 범주형 데이터 칼럼
object_columns = pre_f_df.select_dtypes(include=['object'])

# 결과 출력
display(object_columns)
```

```python
# object 타입을 가진 칼럼들을 제외한 칼럼들 선택
pre_f_num_df = pre_f_df[pre_f_df.columns[~pre_f_df.columns.isin(object_columns)]]
```

```python
pre_f_num_df.loc[:, 'Target'] = pre_f_num_df.Value
```

```python
pre_f_num_df = pre_f_num_df.drop(labels=['Value'],axis=1)
```

```python
pre_f_num_df.corr()['Target'].sort_values(ascending=False)[1:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = pre_f_num_df.iloc[:, :-1], pre_f_num_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
    # print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, R2))

    # 데이터프레임 생성
    evaluation_df = pd.DataFrame({
        'MSE':[MSE], 
        'RMSE':[RMSE], 
        'MSLE':[MSLE],
        'RMSLE':[RMSLE],
        'R2':[R2],
    })

    return evaluation_df
```

```python
prediction = l_r.predict(X_test)
first_test = get_evaluation(y_test, prediction)
```
