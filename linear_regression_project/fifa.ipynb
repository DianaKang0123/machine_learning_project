```python
def calc(currency_string):
    currency_string = currency_string.replace('¢æ', '')  # 통화 기호 제거
    if 'M' in currency_string:
        return float(currency_string.replace('M', '')) * 100000  # M일 경우 백만을 곱함
    elif 'K' in currency_string:
        return float(currency_string.replace('K', '')) * 1000  # K일 경우 천을 곱함
    else:
        return float(currency_string)
```

```python
def height(height):
    # 피트와 인치를 분리
    feet, inches = map(int, height.split("'"))
    # 피트를 센티미터로 변환
    feet_cm = feet * 30.48
    # 인치를 센티미터로 변환
    inches_cm = inches * 2.54
    # 두 값의 합을 반환
    return feet_cm + inches_cm
```

```python
pre_f_df['Height'] = pre_f_df['Height'].apply(height).astype(float)
```

```python
pre_f_df.Height.info()
```

```python
def remove_unit(weight):
    # 숫자만 담을 변수 초기화
    numbers = ''
    # 문자열을 순회하면서 숫자인 부분만 추출
    for char in weight:
        if char.isdigit():
            numbers += char
    return numbers
```

```python
pre_f_df['Weight'] = pre_f_df['Weight'].apply(remove_unit).astype(float)
```

```python
def work_rating(Work_Rate):
    if Work_Rate == 'High/ High':
        return 8
    elif Work_Rate == 'High/ Medium':
        return 7
    elif Work_Rate == 'High/ Low':
        return 6
    elif Work_Rate == 'Medium/ High':
        return 5
    elif Work_Rate == 'Medium/ Medium':
        return 4
    elif Work_Rate == 'Medium/ Low':
        return 3
    elif Work_Rate == 'Low/ High':
        return 2
    elif Work_Rate == 'Low/ Medium':
        return 1
    elif Work_Rate == 'Low/ Low':
        return 0
    else:
        return None
```

```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

features, targets = pca_f_df.iloc[:, :-1], pca_f_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
    train_test_split(features, targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# 데이터를 Tensor로 변환
X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).view(-1, 1)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).view(-1, 1)
X_val = torch.FloatTensor(X_val.values)
y_val = torch.FloatTensor(y_val.values).view(-1, 1)

# 모델 파라미터 초기화
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# 옵티마이저 정의
optimizer = optim.SGD([W, b], lr=1e-17)

# 에포크 설정
epochs = 300000

# 손실 기록 리스트 초기화
train_losses, val_losses, test_losses = [], [], []

# 훈련 루프
for epoch in range(1, epochs + 1):
    # 훈련 데이터에 대한 예측값 및 손실 계산
    H_train = W * X_train + b
    train_loss = torch.mean((y_train - H_train) ** 2)
    
    # 검증 데이터에 대한 예측값 및 손실 계산
    H_val = W * X_val + b
    val_loss = torch.mean((y_val - H_val) ** 2)

    H_test = W * X_test + b
    test_loss = torch.mean((y_test - H_test) ** 2)
    
    # 옵티마이저 초기화
    optimizer.zero_grad()
    
    # 역전파 및 파라미터 업데이트
    train_loss.backward()
    val_loss.backward()
    test_loss.backward()
    optimizer.step()

    # 손실 기록
    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())
    test_losses.append(test_loss.item())
```

```python
# 손실 그래프 출력
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
