```python
def binary_encode(column):
    if column == 'No':
        return 0
    elif column == 'Yes':
        return 1
```

```python
import pandas as pd

def missing_values(series):
    # 결측치 개수 계산
    missing_count = series.isnull().sum()
    
    # 전체 행 수에 대한 결측치 비율 계산
    missing_ratio = (missing_count / len(series)) * 100
    
    # 결과를 시리즈로 반환
    result_series = pd.Series({'Missing Count': missing_count, 'Missing Ratio (%)': missing_ratio})
    
    return result_series
```

```python
missing_values_df = w_df.apply(missing_values)
display(missing_values_df.T)
```

```python
pre_w_df = w_df.copy()
```

```python
pre_w_df
```

```python
missing_values_df = pre_w_df.apply(missing_values)
display(missing_values_df.T)
```

```python
columns = ['Evaporation','Sunshine','Cloud9am','Cloud3pm']

for column in columns:
    pre_w_df.loc[:,column] = pre_w_df[column].fillna(pre_w_df[column].median())
```

```python
missing_values_df = pre_w_df.apply(missing_values)
display(missing_values_df.T)
```

```python
pre_w_df = pre_w_df.dropna()
```

```python
missing_values_df = pre_w_df.apply(missing_values)
display(missing_values_df.T)
```

```python
# 범주형 데이터 칼럼
object_columns = pre_w_df.select_dtypes(include=['object'])

# 결과 출력
display(object_columns)
```

```python
# object 타입을 가진 칼럼들을 제외한 칼럼들 선택
pre_w_num_df = pre_w_df[pre_w_df.columns[~pre_w_df.columns.isin(object_columns)]]
```

```python
pre_w_num_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

features, targets = pre_w_num_df.iloc[:, :-1], pre_w_num_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
train_test_split(X_train, y_train, stratify=y_train, test_size=0.2, random_state=124)

logistic_r = LogisticRegression(solver='liblinear', penalty='l2', C=1, random_state=124)
logistic_r.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction)
    recall = recall_score(y_test , prediction)
    f1 = f1_score(y_test, prediction)
    auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()


    evaluation_df = pd.DataFrame({
        'accuracy': [accuracy],
        'precision': [precision],
        'recall': [recall],
        'f1': [f1],
        'roc_auc': [auc]
    })
    return evaluation_df
```

```python
prediction = logistic_r.predict(X_val)
first_val = get_evaluation(y_val, prediction, logistic_r, X_val)
```

```python
prediction = logistic_r.predict(X_test)
first_test = get_evaluation(y_test, prediction, logistic_r, X_test)
```

```python
import torch
import numpy as np
from torch.nn import Sequential, Linear, Sigmoid
from torch.nn.functional import binary_cross_entropy
from sklearn.model_selection import train_test_split
from torch.optim import SGD

torch.manual_seed(124)

features, targets = under_w_num_df.iloc[:, :-1], under_w_num_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
    train_test_split(features, targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# 데이터를 Tensor로 변환
X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).view(-1, 1)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).view(-1, 1)
X_val = torch.FloatTensor(X_val.values)
y_val = torch.FloatTensor(y_val.values).view(-1, 1)

logistic_r = Sequential(
    Linear(11, 1), # 13개 input을 통해 결과 출력
    Sigmoid(),
)

optimizer = SGD(logistic_r.parameters(), lr=1e-7)

epochs = 1000000

for epoch in range(1, epochs + 1):
    H = logistic_r(X_train)

    loss = binary_cross_entropy(H, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100000 == 0:
        print(f'Epoch: {epoch}/{epochs}')
        for i, w in enumerate(list(logistic_r.parameters())[0][0]):
            print(f'W{i + 1}: {np.round(w.item(), 4)}', end=', ')
        print(f'b: {np.round(list(logistic_r.parameters())[1].item())}\nLoss: {np.round(loss.item(), 4)}')
        print('=' * 90)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction)
    recall = recall_score(y_test , prediction)
    f1 = f1_score(y_test, prediction)
    auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

features, targets = under_w_num_df.iloc[:, :-1], under_w_num_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
    train_test_split(features, targets, test_size=0.2, random_state=124)

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# 데이터를 Tensor로 변환
X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).view(-1, 1)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).view(-1, 1)
X_val = torch.FloatTensor(X_val.values)
y_val = torch.FloatTensor(y_val.values).view(-1, 1)

# 모델 파라미터 초기화
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# 옵티마이저 정의
optimizer = optim.SGD([W, b], lr=1e-7)

# 에포크 설정
epochs = 100000

# 손실 기록 리스트 초기화
train_losses, val_losses, test_losses = [], [], []

# 훈련 루프
for epoch in range(1, epochs + 1):
    # 훈련 데이터에 대한 예측값 및 손실 계산
    H_train = W * X_train + b
    train_loss = torch.mean((y_train - H_train) ** 2)
    
    # 검증 데이터에 대한 예측값 및 손실 계산
    H_val = W * X_val + b
    val_loss = torch.mean((y_val - H_val) ** 2)

    H_test = W * X_test + b
    test_loss = torch.mean((y_test - H_test) ** 2)
    
    # 옵티마이저 초기화
    optimizer.zero_grad()
    
    # 역전파 및 파라미터 업데이트
    train_loss.backward()
    val_loss.backward()
    test_loss.backward()
    optimizer.step()

    # 손실 기록
    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())
    test_losses.append(test_loss.item())
```

```python
# 손실 그래프 출력
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
